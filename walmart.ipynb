```python
import requests
from bs4 import BeautifulSoup

def scrape_walmart_products(pages=2):

    base_url = "https://www.walmart.com/search/?query=laptop&page="
    all_products = []

    print("Starting Walmart scraping...\n")

    for page in range(1, pages + 1):
        print(f"Fetching page {page}...")
        url = f"{base_url}{page}"
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})


        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')


            products = soup.find_all('div', {'data-testid': 'product-container'})
            if not products:
                print(f"No products found on page {page}. Check the website structure.")

            for product in products:
                try:

                    name_element = product.find('a', {'class': 'absolute w-100 h-100 z-1'})
                    price_element = product.find('div', {'class': 'price-main'})

                    if name_element and price_element:
                        name = name_element.get_text(strip=True)
                        price = price_element.get_text(strip=True)
                        link = "https://www.walmart.com" + name_element['href']


                        all_products.append({
                            'name': name,
                            'price': price,
                            'link': link
                        })


                        print(f"Found: {name} - {price}")
                    else:
                        print(f"Incomplete product data on page {page}.")
                except Exception as e:
                    print(f"Error processing a product on page {page}: {e}")
        else:
            print(f"Failed to fetch page {page}: {response.status_code}")

    print("\nScraping completed!\n")
    return all_products


if __name__ == "__main__":
    print("Welcome to the Walmart Scraper!\n")
    num_pages = 2
    products = scrape_walmart_products(pages=num_pages)

    print("\nHere are the products we found:\n")
    for idx, product in enumerate(products, start=1):
        print(f"{idx}. {product['name']} - {product['price']}")
        print(f"   Link: {product['link']}\n")

    print("\nThank you for using the scraper! Have a great day!")
```
